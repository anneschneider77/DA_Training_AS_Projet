## ajout du calcul de la distance entre la commune et le chef_lieu de région, ou n'importe qu'elle autre commune

```python
import numpy as np

def spherical_dist(pos1, pos2, r=3958.75 * 1.609):
    pos1 = pos1 * np.pi / 180
    pos2 = pos2 * np.pi / 180
    cos_lat1 = np.cos(pos1[..., 0])
    cos_lat2 = np.cos(pos2[..., 0])
    cos_lat_d = np.cos(pos1[..., 0] - pos2[..., 0])
    cos_lon_d = np.cos(pos1[..., 1] - pos2[..., 1])
    return r * np.arccos(cos_lat_d - cos_lat1 * cos_lat2 * (1 - cos_lon_d))


positions1 = fusion_1[['latitude', 'longitude']].values
positions2 = fusion_1[fusion_1['nom_commune'] == 'Paris'][['latitude', 'longitude']].values[0, :]


distances_to_paris = spherical_dist(positions1, positions2)
```

```python
# renommage des colonnes
population['SEXE'] = population['SEXE'].replace({1: 'homme', 2: 'femme'})
population['MOCO'] = population['MOCO'].replace({v: 'moco_'+ str(v) for v in population['MOCO'].unique()})
population['AGEQ80_17'] = population['AGEQ80_17'].replace({v: 'age_'+ str(v) for v in population['AGEQ80_17'].unique()})
population.head()
```

```python

population = population.replace(to_replace = ['moco_moco_11', 'moco_moco_12', 'moco_moco_21', 'moco_moco_22',
       'moco_moco_23', 'moco_moco_31', 'moco_moco_32'], value = ['Enf av 2parents', 'Enf av 1parent', 'Coupl ss Enf', 'Coupl av Enf',
       'Cel av Enf', 'Adulte + du foyer', 'Adulte Seul'])



population = population.replace(to_replace = ['age_age_0', 'age_age_5', 'age_age_10', 'age_age_15', 'age_age_20',
       'age_age_25', 'age_age_30', 'age_age_35', 'age_age_40',
       'age_age_45', 'age_age_50', 'age_age_55', 'age_age_60',
       'age_age_65', 'age_age_70', 'age_age_75', 'age_age_80'], 
                                value = ['age_0', 'age_5', 'age_10', 'age_15', 'age_20',
       'age_25', 'age_30', 'age_35', 'age_40',
       'age_45', 'age_50', 'age_55', 'age_60',
       'age_65', 'age_70', 'age_75', 'age_80'])

population['AGEQ80_17'].unique()
```

```python
dfs = []

# parcours les colonnes
for col in ['MOCO', 'AGEQ80_17', 'SEXE']:
    
    # parcours les valeurs uniques des colonnes
    for value in population[col].unique():
        
        df_temp = population[population[col] == value].groupby('code_insee').agg({'NB': sum})
        
        df_temp[value] = df_temp['NB']
        df_temp = df_temp.drop(['NB'], axis=1)
        
        dfs.append(df_temp)
```

```python
# pd.contact permet de concaténer plusieurs DF entre eux () tous les df dans dfs ont les mêmes indices), 
# axis 1 , en 'ligne', ignore_index = False pour garder code_insee en index
population_total = pd.concat(
    dfs, 
    axis=1, 
    ignore_index=False
)
population_total.head()
```
