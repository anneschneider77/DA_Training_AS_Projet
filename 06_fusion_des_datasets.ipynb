## Récupération des codes geo

```python
cod_geos = {
    'ent': entreprises['codgeo'].unique(),
    'sal': salary['codgeo'].unique(),
    'geo': geography['codgeo'].unique(),
    'cons':conso['codgeo'].unique(),
    'pop': population['codgeo'].unique()
}
```

```python
# vérification de l'unicité
print('entreprises: ', entreprises[['codgeo']].duplicated().sum())
print('salaires: ', salary[['codgeo']].duplicated().sum())
print('geographie: ', geography[['codgeo']].duplicated().sum())
print('consommation: ', conso[['codgeo']].duplicated().sum())
print('population: ', population[['codgeo']].duplicated().sum())
```

```python
# on met salary de côté pour l'instant

print('salaires: ', len(cod_geos['sal']))
print('entreprises: ', len(cod_geos['ent']))
print('population: ', len(cod_geos['pop']))
print('geographie: ', len(cod_geos['geo']))
print('consommation: ', len(cod_geos['cons']))
```

## Méthodologie

On va se servir de `geography` comme base. On va regarder pour chaque jeu de données, les codes manquants et les codes en plus

```python
differences = {}

for key1 in cod_geos:
    for key2 in cod_geos:
        diff_temp = set(cod_geos[key1]).difference(cod_geos[key2])
        differences['{}-{}'.format(key1, key2)] = {
            'nb': len(diff_temp),
            'diff': diff_temp
        }
```

### Fusion de géographie et entreprises

```python
geo_ent = pd.merge(
    left=geography,
    right=entreprises,
    on=['codgeo'],
    how='left'
)
```

```python
geo_ent.head()
```

```python
col = 'nb_ent'
geo_ent[['nom_region', col]].groupby(['nom_region']).agg({col: lambda x: x.isna().mean()})
```

```python
geo_ent.isna().sum()
```

```python
# geo_ent = geo_ent.dropna()
```

```python
geo_ent_pop = pd.merge(
    left=geo_ent,
    right=population,
    on=['codgeo'],
    how='left'
)

geo_ent_pop.head()
```

```python
geo_ent_pop.isna().sum()
```

```python
population[population['codgeo'].isin(differences['pop-geo']['diff'])].describe()
```

```python
population[population['codgeo'].isin(differences['pop-geo']['diff'])].sort_values('pop_total').tail(20)
```

```python
geography[geography['nom_commune'] == 'Courbevoie']
```

### Fusion avec conso

```python
geo_ent_pop_conso = pd.merge(
    left=geo_ent_pop,
    right=conso,
    how='left',
    on=['codgeo']
)

geo_ent_pop_conso.head()
```
